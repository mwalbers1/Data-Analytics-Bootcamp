### Creating Bubble chart using MATLAB approach

Bubble charts are useful when presenting financial, population, and weather data because you can add a third and fourth factor to convey more information.

The first two factors are the x- and y-axes data, which we have been using quite frequently. By changing the "dot" into a "bubble," we are adding a third factor: size. If there is more than one dataset that uses the same axes, we can change the color of each marker for each dataset, which will add a fourth factor: color.

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# list comprehension
# new_list = [expression for item in list if conditional]
```

## 5.1.8 Create Pie Charts

Pie charts provide a way to visualize percentages

### Create a Pie Chart Using the MATLAB method

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))

plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')
plt.show()
```

### Create Pie chart using the object-oriented interface

```python
colors = ["skyblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,labels=x_axis, autopct="%.1f%%", shadow=True, startangle=95, counterclock=False, colors=colors, explode=explode_values)
plt.show()
```

## 5.3.1 Create dataframes for each type of city

```python
# Create the Urban DataFrame
urban_cities_df = pyber_data_df[pyber_data_df["type"]=="Urban"]
urban_cities_df.head()
```

```python
suburban_cities_df = pyber_data_df[pyber_data_df["type"]=="Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"]=="Rural"]
```

```python
rural_cities_df.head()
```

## 5.3.2 Get the number of Rides for Each city type

```python
# Get the number of rides for urban cities
urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"]
urban_ride_count.head()
```

```python
suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"]
suburban_ride_count
```

```python
rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"]
rural_ride_count
```

## Get the Average Fare for Each City Type

```python
urban_avg_fare = urban_cities_df.groupby(["city"]).mean()["fare"]
urban_avg_fare.head()
```

```python
# Get average fare for each city in the suburban and rural cities
suburban_avg_fare = suburban_cities_df.groupby(["city"]).mean()["fare"]
rural_avg_fare = rural_cities_df.groupby(["city"]).mean()["fare"]
```
