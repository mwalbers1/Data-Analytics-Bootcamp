## 5.3.5 Create Bubble Charts

```python
# define function to create bubble chart
def create_bubble_chart(ride_count, avg_fare, driver_count, color, label):
    """
    create bubble chart for args passed in
    
    args:
        ride_count: ride count series
        avg_fare: average fare series
        driver_count: driver count series
        color: color for label
        label: label 
    """
    plt.scatter(ride_count, avg_fare, s=10*driver_count, c=color, \
            edgecolor="black", linewidths=1, alpha=0.8, label=label )
    plt.title("Pyber Ride-Sharing Data (2019)")
    plt.ylabel("Average Fare ($)")
    plt.xlabel("Total Number of Rides (Per City)")
    plt.grid(True)
    plt.legend()
```

```python
# Build scatter plot for urban cities
plt.scatter(urban_ride_count, urban_avg_fare, s=10*urban_driver_count, c="coral", \
            edgecolor="black", linewidths=1, alpha=0.8, label="Urban" )
plt.title("Pyber Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
#build scatter plot for rural cities
create_bubble_chart(rural_ride_count, rural_avg_fare, rural_driver_count, color="gold", label="rural")
```

```python
#build scatter plot for suburban cities
create_bubble_chart(suburban_ride_count, suburban_avg_fare, suburban_driver_count, color="skyblue", label="rural")
```

```python
# Add the scatter charts for each type of city.
# make the urban markers coral and 20% transparent.
plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=10*urban_driver_count, c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=10*suburban_driver_count, c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")

# Incorporate the other graph properties
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.xlabel("Total Number of Rides (Per City)", fontsize=12)
plt.grid(True)

# Add legend
lgnd = plt.legend(fontsize="12", mode="Expanded", scatterpoints=1, loc="best", title="City Types")

lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

# Text label about circle size
plt.text(42, 35, "Note: Circle sizes correlates with driver count per city", fontsize=12)

# Show the plot
plt.show()
```

## 5.4.4 Create Box-and-Whisker Plots

```python
def create_box_plot(label, title, series, ytitle=None):
    """
    Plot box plot for series passed into function
    
    args:
        label: x axis lable
        series: series object to plot box plot for
    """
    
    x_labels = [label]
    fig, ax = plt.subplots(figsize=(6,7))
    ax.boxplot(series, labels=x_labels)

    # Add the title, y-axis label and grid
    bp_title = "Ride Count Data (2019)" if title==None else title
    y_title = "Number of Rides" if ytitle==None else ytitle
    
    ax.set_title(bp_title)
    ax.set_ylabel(y_title)
    ax.set_yticks(np.arange(0,90, step=5.0))
    ax.grid()
    plt.show()
```
