## 4.8.7 Format the DataFrame

```python
# define method to return formatted school summary dataframe
def format_school_summary_dataframe(df):
    """
    Format numeric and float columns of school summary dataframe
    
    args: 
    df : school summary dataframe
    
    returns: formatted school summary dataframe
    """
    per_school_summary_df_formatted = pd.DataFrame(
        {
            "School Type":df["School Type"],
            "Total Students":df["Total Students"].map("{:,}".format),
            "Total School Budget":df["Total School Budget"].map("${:,.2f}".format),
            "Per Student Budget":df["Per Student Budget"].map("${:,.2f}".format),
            "Average Math Score":df["Average Math Score"].map("{:.1f}".format),     
            "Average Reading Score":df["Average Reading Score"].map("{:.1f}".format),
            "% Passing Math":df["% Passing Math"].map("{:.1f}".format),
            "% Passing Reading":df["% Passing Reading"].map("{:.1f}".format),
            "% Overall Passing":df["% Overall Passing"].map("{:.1f}".format) 
        }
    )
    return per_school_summary_df_formatted
```

- "Small" schools have fewer than 1,000 students
